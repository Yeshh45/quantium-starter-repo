```python
import pandas as pd
```

```python
f1 = pd.read_csv('data/daily_sales_data_0.csv')
```

```python
f1.head(5)
```

```python
f1[f1['product']=='pink morsel']
```

```python
pm_f1 = f1[f1['product']=='pink morsel']
```

```python
pm_f1['price'] = (
    pm_f1['price']
    .astype(str)
    .str.replace(r'[^0-9.]', '', regex=True)
    .astype(float)
)
```

```python
pm_f1['sales'] = pm_f1['price']*pm_f1['quantity']
```

```python
filtered_f1 = pm_f1[['product', 'sales', 'date', 'region']]
```

```python
filtered_f1.head()
```

```python
filtered_f1['product'].count()
```

```python
filtered_f1.head()
```

```python
f2 = pd.read_csv('data/daily_sales_data_1.csv')
```

```python
f3 = pd.read_csv('data/daily_sales_data_2.csv')
```

```python
pm_f2 = f2[f2['product']=='pink morsel']
pm_f2['price'] = (
    pm_f2['price']
    .astype(str)
    .str.replace(r'[^0-9.]', '', regex=True)
    .astype(float)
)
pm_f2['sales'] = pm_f2['price']*pm_f2['quantity']
filtered_f2 = pm_f2[['product', 'sales', 'date', 'region']]
```

```python
pm_f3 = f3[f3['product']=='pink morsel']
pm_f3['price'] = (
    pm_f3['price']
    .astype(str)
    .str.replace(r'[^0-9.]', '', regex=True)
    .astype(float)
)
pm_f3['sales'] = pm_f3['price']*pm_f3['quantity']
filtered_f3 = pm_f3[['product', 'sales', 'date', 'region']]
```

```python
f = pd.concat([filtered_f1, filtered_f2, filtered_f3], ignore_index=True)
```

```python
f.head()
```

```python
final_data = f[['sales', 'date', 'region']]
```

```python
final_data.head()
```

```python
final_data.to_csv('final_data.csv', index=False)
```
